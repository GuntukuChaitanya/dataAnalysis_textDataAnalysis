```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
comments=pd.read_csv('GBcomments.csv',error_bad_lines=False)
```

```python
comments
```

```python
from textblob import TextBlob
TextBlob('Its more accurate to call it the M+ (1000) be...	').sentiment.polarity
```

```python
comments.isna().sum()
```

```python
comments.dropna(inplace=True)
```

```python
polarity=[]
for i in comments['comment_text']:
    polarity.append(TextBlob(i).sentiment.polarity)
```

```python
comments['polarity']=polarity
```

```python
comments
```

```python
comments.head(20)
```

```python
comments_positive=comments[comments['polarity']==1]
```

```python
comments_positive.shape
```

```python
comments_positive
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
stopwords=set(STOPWORDS)
```

```python
total_positive_comments=' '.join(comments_positive['comment_text'])
```

```python
wordcloud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(total_positive_comments)
```

```python
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```

```python
comments_negative=comments[comments['polarity']==-1]
total_negative_comments=' '.join(comments_negative['comment_text'])
wordcloud=WordCloud(width=1000,height=500,stopwords=stopwords).generate(total_negative_comments)
plt.figure(figsize=(15,5))
plt.imshow(wordcloud)
plt.axis('off')
```
